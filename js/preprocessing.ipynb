```python
import pandas as pd
import scipy
import numpy as np
```

```python
data = pd.read_csv("../data/heatmap_data.csv")
```

```python
data
```

```python
column_name = data.columns
result = []
for column_one in column_name[1:]:
    for column_two in column_name[1:]:
        list_one = data[column_one]
        list_two = data[column_two]
        list_one_res = []
        list_two_res = []
        for i in range(len(list_one)):
            if (list_one[i]!= "NAN" and list_two[i]!= "NAN"):
                list_one_res.append(float(list_one[i].replace(',','.')))
                list_two_res.append(float(list_two[i].replace(',','.')))
        list_one_res = np.array(list_one_res)
        list_two_res = np.array(list_two_res)
        sort_order = np.argsort(list_one_res)
        list_one_res = list_one_res[sort_order]
        list_two_res = list_two_res[sort_order]
        
        res = scipy.stats.kendalltau(list_one_res, list_two_res)
        result.append([column_one,column_two,res.correlation])

```

```python
result
```

```python
df = pd.DataFrame(result, columns = ['Column_one', 'Column_two', 'Correlation']) 
```

```python
df.to_csv('../data/correlation_matrix.csv', index = False)
```
